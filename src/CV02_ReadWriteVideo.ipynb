```python
import cv2 as cv
```

# Read Display Save videos using cameras

```python
cap = cv.VideoCapture(0) # here we can provide any video clip or we can provide our camera index
# the default camera index -> 0

# now we will create a while loop in order to capture the frame continuously
while(True):
    ret, frame = cap.read() # ret will store True or False according to frame availibility
    # and frame will store thr capture frame
    
    cv.imshow('frame', frame)
    if cv.waitKey(1) & 0xFF == ord('q'): 
        break

cap.release()
cv.destroyAllWindows()
```

### Converting Live stream in gray scalling
You can get more PROP ID here -> https://docs.opencv.org/4.0.0/d4/d15/group__videoio__flags__base.html#gaeb8dd9c89c10a5c63c139bf7c4f5704d

```python
cap = cv.VideoCapture(0)
print(cap.isOpened())
while(cap.isOpened()): # this will only work when we get a valid video file or valid index of available camera
    ret, frame = cap.read()
    
    print(cap.get(cv.CAP_PROP_FRAME_WIDTH))
    print(cap.get(cv.CAP_PROP_FRAME_HEIGHT))
    
    gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY) # first parameter is the frame which is need to conert in the gray scalling, and second is the converson
    cv.imshow('frame', gray)
    if cv.waitKey(1) & 0xFF == ord('q'): 
        break

cap.release()
cv.destroyAllWindows()
```
