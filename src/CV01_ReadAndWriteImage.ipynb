# Display Image

#### Gray Scalling

```python
image = cv.imread('CV01_lena.jpg', 0)
```

```python
cv.imshow('Image', image)
# it opens a window but for one milli sec
# and also window will remain open if we run the code in notebook
# window will get stuck bcz here it wont have the termination commond for window, which we need to give manually here in notebook
```

```python
cv.imshow('Image', image)
cv.waitKey(6000)
cv.destroyAllWindows() # this commond is used to terminate all the open window
```

```python
cv.imshow('Image', image)
cv.waitKey(0) # now the open window will not get disappear any number of seconds, instead we need to terinate
cv.destroyAllWindows()
```
